* Here, We can see outliers in 6 numerical columns. They are Totalworkexperienceinyears, Experienceyearsatthiscompany,trainingtimelastyear,experienceincurrentrole, yearssincelastpromotion,yearswithcurrmanager

```python
# Identify numerical columns with outliers
num_out=data[['TotalWorkExperienceInYears', 'ExperienceYearsAtThisCompany',
              'TrainingTimesLastYear','ExperienceYearsInCurrentRole','YearsSinceLastPromotion',
              'YearsWithCurrManager']]
```

```python
# Calculate the median for each column
median_values = num_out.median()

# Define a function to replace outliers with the median
def replace_outliers_with_median(x):
    median = median_values[x.name]
    x[x < x.quantile(0.25)] = median
    x[x > x.quantile(0.75)] = median
    return x

# Apply the function to replace outliers with the median
num_out = num_out.apply(replace_outliers_with_median, axis=0)
```

```python
# Create a figure to display box plots
plt.figure(figsize=(12, 6))

# Plot box plots for each numerical column before and after outlier treatment
for i, column in enumerate(num_out.columns):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(x=num_out[column])
    plt.title(f"{column} - After Outlier Handling")

plt.tight_layout()
plt.show()
```

* Even after imputing median , total work experience and years with current manager still have outliers.

### BALANCING THE TARGET FEATURE
 * SMOTE (synthetic minority oversampling technique) is one of the most commonly used oversampling methods to solve the imbalance problem. It aims to balance class distribution by randomly increasing minority class examples by replicating them. SMOTE synthesises new minority instances between existing minority instances.



```python
from collections import Counter
from imblearn.over_sampling import SMOTE #SMOTE(synthetic minority oversampling techinque)
sm = SMOTE() # obeject creation
print("unbalanced data   :  ",Counter(y))
X_sm,y_sm = sm.fit_resample(X,y)
print("balanced data:    :",Counter(y_sm))
```

* Now the target feature is balanced.

## Spliting Training and Testing data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_sm,y_sm,random_state=42,test_size=0.20)
```

## PRINCIPLE COMPONENT ANALYSIS

   Principal component analysis (PCA) is a popular technique for analyzing large datasets containing a high number of dimensions/features per observation, increasing the interpretability of data while preserving the maximum amount of information, and enabling the visualization of multidimensional data. Formally, PCA is a statistical technique for reducing the dimensionality of a dataset.

```python
# Getting optimal value of pca
from sklearn.decomposition import PCA
pca = PCA() 
principlecomponents = pca.fit_transform(data) # Fitting the data
plt.figure(figsize=(10,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_),marker='*',color='k') #plot the variance ratio
plt.xlabel('Number of components')
plt.ylabel('Variance(%)') # for each componant
plt.title('Explained Variance',fontsize=20)
plt.show()
```

* From the above PCA graph, we can understand the number of components can be 25 because 25 components have the elbow points. But the original data have 28 features.  
* We have opted to utilize the original dataset, comprising 28 features, instead of employing the 25 important components derived from PCA.
